```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
root_path = os.path.join(os.path.dirname(os.getcwd()), "FuelPoverty_data", "HeatingProfileandDemand")
```

```python
heat_profile_df = pd.read_csv(os.path.join(root_path, "heat_tech_profiles.csv"))
heat_profile_df
```

```python
heat_profile_df["index"]
```

```python
plt.figure(figsize=[6,2])
plt.plot(heat_profile_df["index"], heat_profile_df["Normalised_ASHP_elec"])
plt.title("Normalised HP Electricity Demand over a year")

# A hacky way of showing a subset of x ticks
dates = heat_profile_df["index"].to_numpy()
plt.ylabel("kWh")
plt.xticks([dates[i] for i in range(0, len(dates), 1500)], [str(dates[i])[:7] for i in range(0, len(dates), 1500)], rotation = 45)

plt.show()
```

```python
plt.figure(figsize=[6,2])
plt.plot(heat_profile_df["index"], heat_profile_df["Normalised_Resistance_heater_elec"])
plt.title("Normalised Resistive Heating Electricity Demand over a year")

# A hacky way of showing a subset of x ticks
dates = heat_profile_df["index"].to_numpy()
plt.ylabel("kWh")
plt.xticks([dates[i] for i in range(0, len(dates), 1500)], [str(dates[i])[:7] for i in range(0, len(dates), 1500)], rotation = 45)

plt.show()
```

```python
plt.figure(figsize=[6,2])
plt.plot(heat_profile_df["index"], heat_profile_df["Normalised_Resistance_heater_heat"])
plt.title("Normalised Resistive Heating Electricity Demand over a year")

# A hacky way of showing a subset of x ticks
dates = heat_profile_df["index"].to_numpy()
plt.ylabel("kWh")
plt.xticks([dates[i] for i in range(0, len(dates), 1500)], [str(dates[i])[:7] for i in range(0, len(dates), 1500)], rotation = 45)

plt.show()
```
